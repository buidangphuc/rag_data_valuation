# Chunking Strategy Comparison
This notebook compares different chunking strategies (Fixed, Recursive, Semantic) and their impact on Data Valuation scores.

```python
from src.rag.chunking.simple import fixed_length_chunker
from src.rag.chunking.recursive import recursive_chunker
from src.rag.chunking.semantic import semantic_chunker

text = "Paris is the capital of France. It is known for the Eiffel Tower. France is in Europe."
fixed = fixed_length_chunker(text)
rec = recursive_chunker(text)
sem = semantic_chunker(text)

print(f'Fixed: {len(fixed)} chunks')
print(f'Recursive: {len(rec)} chunks')
print(f'Semantic: {len(sem)} chunks')
```
